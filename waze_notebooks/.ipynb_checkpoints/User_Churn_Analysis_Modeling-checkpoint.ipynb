#### Apply descriptive statistics and hypothesis testing to analyze the difference in the mean amount of rides between iPhone users and Android users.

```python
from scipy import stats
```

```python
# Encode 'device', create a new column'device_new' 
waze['device_new'] = np.where(waze['device'] == 'Android',2,1)
waze.head()
```

```python
# Investigate relationship between device type and the number of drives.
waze.groupby('device_new')['drives'].mean()
```

Based on the averages shown, it appears that drivers who use an iPhone device have a higher number of drives on average. However, this difference might arise from random sampling, rather than being a true difference in the number of drives. To assess whether the difference is statistically significant, I will conduct a two-sample t-test for two independent samples(Andriod users vs. iphone users).

* Step 1: State the null hypothesis and the alternative hypothesis
* Step 2: Choose a signficance level: 5%
* Step 3: Find the p-value
* Step 4: Reject or fail to reject the null hypothesis

$𝐻_0$: there is no difference between the two types of devices regarding the drives. 
$H_A$: there is difference between the two types of devices regarding the drives.

```python
drives_iphone = waze[waze['device_new'] == 1]['drives']
drives_android = waze[waze['device_new'] == 2]['drives']
stats.ttest_ind(a=drives_iphone,b=drives_android,equal_var=False)
```

pvalue > significance level: 0.05(i.e.5%), therefore fail to reject the null hypothesis, meaning there is not statistically significant difference in avg drives between iphone and android users, the difference shown can be by chance.

#### 2. Handling outliers

The box plots from the previous section indicated that many of these variables have outliers. These outliers do not seem to be data entry errors; they are present because of the right-skewed distributions.

Because I will use the data to build logistic regression model, it may be useful to impute outlying data with more reasonable values. I will set a threshold based on a percentile of the distribution.

For this analysis, I will impute the outliers values with the **95th percentile** threshold.

```python
# Creat a helper function to imput outliers
def outlier_imputer(column_name, percentile):
    # Calculate threshold
    threshold = waze[column_name].quantile(percentile)
    # Impute threshold for values > than threshold
    waze.loc[waze[column_name] > threshold, column_name] = threshold
    print('{:>25} | percentile: {} | threshold: {}'.format(column_name, percentile, threshold))
```

```python
# Impute outliers
for column in ['sessions', 'drives', 'total_sessions', 'total_navigations_fav1',
               'total_navigations_fav2', 'driven_km_drives', 'duration_minutes_drives']:
    outlier_imputer(column,0.95)
```

```python
# Define a helper function to output all of the scores of the models
def make_results(model_name:str, model_object, metric:str):
    '''
        model_name (string): what the model is called in the output table
        model_object: a fit GridSearchCV model object
        metric (string): precision, recall, f1, or accuracy

    Returns a pandas dataframe with the F1, recall, precision, and accuracy scores
    for the model with he best mean 'metric' score across all validation folds.
    '''

    # Create dictionary that maps input metric to actual metric name in GridSearchCV
    metric_dict = {'precision': 'mean_test_precision',
                   'recall': 'mean_test_recall',
                   'f1': 'mean_test_f1',
                   'accuracy': 'mean_test_accuracy',
                   }

    # Get all the results from the CV and put them in a df
    cv_results = pd.DataFrame(model_object.cv_results_)

    # Isolate the row of the df with the max(metric) score,idxmax()returns a Series 
    # with the index of the maximum value for each row.
    best_estimator_results = cv_results.iloc[cv_results[metric_dict[metric]].idxmax(), :]

    # Extract accuracy, precision, recall, and f1 score from that row
    f1 = best_estimator_results.mean_test_f1
    recall = best_estimator_results.mean_test_recall
    precision = best_estimator_results.mean_test_precision
    accuracy = best_estimator_results.mean_test_accuracy

    # Create table of results
    table = pd.DataFrame({'model': [model_name],
                          'precision': [precision],
                          'recall': [recall],
                          'F1': [f1],
                          'accuracy': [accuracy],
                          },
                         )

    return table
```

```python
# Define a helper function to generate a table of scores from the predictions on the validation data.
def get_test_scores(model_name:str,preds,y_test_data):
    accuracy = accuracy_score(y_test_data, preds)
    precision = precision_score(y_test_data, preds)
    recall = recall_score(y_test_data, preds)
    f1 = f1_score(y_test_data, preds)

    table = pd.DataFrame({'model': [model_name],
                          'precision': [precision],
                          'recall': [recall],
                          'F1': [f1],
                          'accuracy': [accuracy]
                          })

    return table
```

```python
# Call the get_test_scores on random forest model
rf_val_scores = get_test_scores('RF val',rf_val_preds,y_val)
```

```python
# Append to the results table
results = pd.concat([results, rf_val_scores], axis=0)
results
```
